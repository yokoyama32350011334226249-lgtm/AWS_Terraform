```python
import json
import boto3

s3 = boto3.client("s3")

bucket = "xxx" # バケット名
prefix = "nikkei-deepar"

obj = s3.get_object(
    Bucket=bucket,
    Key=f"{prefix}/output/predict.json.out"
)

result = json.loads(obj["Body"].read())
forecast = result["quantiles"]["0.4"]

print(forecast)

import matplotlib.pyplot as plt
import pandas as pd
import json

df = pd.read_csv("nikkei.csv", encoding="cp932")
actual = df["close"].tail(30)
forecast_series = forecast

plt.figure(figsize=(10,5))
plt.plot(actual.values, label="actual")
plt.plot(
    range(len(actual), len(actual) + len(forecast_series)),
    forecast_series,
    label="forecast"
)
plt.legend()
plt.show()

df2 = pd.read_csv("nikkei-2.csv", encoding="cp932")
actual2 = df2["close"].tail(35)

plt.figure(figsize=(10,5))
plt.plot(actual2.values, label="actual2")
plt.legend()
plt.show()
```
